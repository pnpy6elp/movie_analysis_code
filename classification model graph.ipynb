# LSTM

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
#plt.figure(dpi=300)
ind = np.arange(6)
width = 0.25
def1 = [0.6141,0.6976,0.5844,0.6008,0.6320,0.6656]
def2 = [0.5586,0.5789,0.5367,0.5609,0.5234,0.5719]
def3 = [0.5406,0.5781,0.5437,0.5546,0.5445,0.5258]
plt.grid(True, axis='y', alpha=0.5)
p1 = plt.bar(ind, def1, width,color='dodgerblue', edgecolor = 'darkblue',alpha=0.5)
p2 = plt.bar(ind+0.25, def2, width, color='darkseagreen', edgecolor = 'darkgreen',alpha=0.5)
p3 = plt.bar(ind+0.5, def3, width, color='orangered', edgecolor = 'darkred',alpha=0.5)
plt.xticks(np.arange(width, 6 + width, 1), ("Assassination","The spy\ngone north","1987","A taxi\ndriver","Intimate\nstranger","Confidential\nassignment"),fontsize=15)
lg = plt.legend((p1[0],p2[0],p3[0]),('Definition 1','Definition 2','Definition 3'),prop={'size': 10},bbox_to_anchor=(0.7, -0.15),fancybox=True, ncol=5)
plt.ylabel("Accuracy")

plt.savefig("./graph/lstm.png",dpi=300,bbox_extra_artists=(lg,), 
            bbox_inches='tight')
plt.show()
```
