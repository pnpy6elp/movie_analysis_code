```python
#공조
import pandas as pd
import numpy
import json
import matplotlib.pyplot as plt

import sklearn.metrics as sk 



df = pd.read_csv("./dataset/gongjo.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
# li : 1은 t개, 0은 ut개 (정답과 크기를 맞추기 위해 설정)
df["correlation"] = df["correlation"].abs()
df = df.reset_index()
df["pred"] = li # column 추가

df_t1 = df[df["label"] == 1] # trust
df_ut1 = df[df["label"] == 0] # distrust
df1 = df_t1[['rating','sentiment','corr','review']]
df2 = df_ut1[['rating','sentiment','corr','review']]


df1.to_csv('./helpfulness/gongjo_t.txt',encoding="utf-8-sig",header=False,index=False) 
df2.to_csv('./helpfulness/gongjo_ut.txt',encoding="utf-8-sig",header=False,index=False)  
print("공조 : ",t," ",ut)

# 암살
df = pd.read_csv("./dataset/assassin.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
# li : 1은 t개, 0은 ut개 (정답과 크기를 맞추기 위해 설정)
df["correlation"] = df["correlation"].abs()
df = df.reset_index()
df["pred"] = li # column 추가

df_t1 = df[df["label"] == 1] # trust
df_ut1 = df[df["label"] == 0] # distrust
df1 = df_t1[['rating','sentiment','corr','review']]
df2 = df_ut1[['rating','sentiment','corr','review']]


df1.to_csv('./helpfulness/assassin_t.txt',encoding="utf-8-sig",header=False,index=False) 
df2.to_csv('./helpfulness/assassin_ut.txt',encoding="utf-8-sig",header=False,index=False)  
print("assassin : ",t," ",ut)

# 1987
df = pd.read_csv("./dataset/1987.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
# li : 1은 t개, 0은 ut개 (정답과 크기를 맞추기 위해 설정)
df["correlation"] = df["correlation"].abs()
df = df.reset_index()
df["pred"] = li # column 추가

df_t1 = df[df["label"] == 1] # trust
df_ut1 = df[df["label"] == 0] # distrust
df1 = df_t1[['rating','sentiment','corr','review']]
df2 = df_ut1[['rating','sentiment','corr','review']]


df1.to_csv('./helpfulness/1987_t.txt',encoding="utf-8-sig",header=False,index=False) 
df2.to_csv('./helpfulness/1987_ut.txt',encoding="utf-8-sig",header=False,index=False)  
print("1987 : ",t," ",ut)

# intistranger
df = pd.read_csv("./dataset/intistranger.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
# li : 1은 t개, 0은 ut개 (정답과 크기를 맞추기 위해 설정)
df["correlation"] = df["correlation"].abs()
df = df.reset_index()
df["pred"] = li # column 추가

df_t1 = df[df["label"] == 1] # trust
df_ut1 = df[df["label"] == 0] # distrust
df1 = df_t1[['rating','sentiment','corr','review']]
df2 = df_ut1[['rating','sentiment','corr','review']]


df1.to_csv('./helpfulness/intistranger_t.txt',encoding="utf-8-sig",header=False,index=False) 
df2.to_csv('./helpfulness/intistranger_ut.txt',encoding="utf-8-sig",header=False,index=False)  
print("완벽한 타인 : ",t," ",ut)

df = pd.read_csv("./dataset/spynorth.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
# li : 1은 t개, 0은 ut개 (정답과 크기를 맞추기 위해 설정)
df["correlation"] = df["correlation"].abs()
df = df.reset_index()
df["pred"] = li # column 추가

df_t1 = df[df["label"] == 1] # trust
df_ut1 = df[df["label"] == 0] # distrust
df1 = df_t1[['rating','sentiment','corr','review']]
df2 = df_ut1[['rating','sentiment','corr','review']]


df1.to_csv('./helpfulness/spynorth_t.txt',encoding="utf-8-sig",header=False,index=False) 
df2.to_csv('./helpfulness/spynorth_ut.txt',encoding="utf-8-sig",header=False,index=False)  
print("spynorth : ",t," ",ut)

df = pd.read_csv("./dataset/taxi.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
# li : 1은 t개, 0은 ut개 (정답과 크기를 맞추기 위해 설정)
df["correlation"] = df["correlation"].abs()
df = df.reset_index()
df["pred"] = li # column 추가

df_t1 = df[df["label"] == 1] # trust
df_ut1 = df[df["label"] == 0] # distrust
df1 = df_t1[['rating','sentiment','corr','review']]
df2 = df_ut1[['rating','sentiment','corr','review']]


df1.to_csv('./helpfulness/taxi_t.txt',encoding="utf-8-sig",header=False,index=False) 
df2.to_csv('./helpfulness/taxi_ut.txt',encoding="utf-8-sig",header=False,index=False)  
print("taxi : ",t," ",ut)
```
