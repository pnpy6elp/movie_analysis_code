```python
#공조
import pandas as pd
import numpy
import json
import matplotlib.pyplot as plt

import sklearn.metrics as sk 



df = pd.read_csv("./dataset/gongjo.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
# li : 1은 t개, 0은 ut개 (정답과 크기를 맞추기 위해 설정)
df["correlation"] = df["correlation"].abs()
df = df.sort_values(by=['rating std'], axis=0, ascending=False) 
df = df.reset_index()
# correlation score을 기준으로 오름차순, 위에서 t번째까지는 pred가 1, 나머지는 untrust이므로 0개 
df["pred"] = li # column 추가
df_a1 = df[df["label"] == 1] # trust
df_a2 = df[df["label"] == 0] # distrust

df_t1 = df[df["pred"] == 1] # trust
df_ut1 = df[df["pred"] == 0] # distrust
df1 = df_t1[['review','rating std','sentiment std','correlation']]
df2 = df_ut1[['review','rating std','sentiment std','correlation']]


df1.to_csv('./definition/definition1_gongjo_t.txt',encoding="utf-8-sig",header=False,index=False) 
df2.to_csv('./definition/definition1_gongjo_ut.txt',encoding="utf-8-sig",header=False,index=False)  
print("공조 : ",t," ",ut)

df = pd.read_csv("./dataset/gongjo.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
# li : 1은 t개, 0은 ut개 (정답과 크기를 맞추기 위해 설정)
df["correlation"] = df["correlation"].abs()
df = df.sort_values(by=['sentiment std'], axis=0, ascending=False) 
df = df.reset_index()
# correlation score을 기준으로 오름차순, 위에서 t번째까지는 pred가 1, 나머지는 untrust이므로 0개 
df["pred"] = li # column 추가
df_t1 = df[df["pred"] == 1] # trust
df_ut1 = df[df["pred"] == 0] # distrust
df1 = df_t1[['review','rating std','sentiment std','correlation']]
df2 = df_ut1[['review','rating std','sentiment std','correlation']]



df1.to_csv('./definition/definition2_gongjo_t.txt',encoding="utf-8-sig",header=False,index=False)  
df2.to_csv('./definition/definition2_gongjo_ut.txt',encoding="utf-8-sig",header=False,index=False) 


df = pd.read_csv("./dataset/gongjo.csv", sep=",")
df = df[df["sympathy"] != df["unsympathy"]]

for index, row in df.iterrows(): 
    if df.loc[index, "sympathy"] > df.loc[index, "unsympathy"]:
        df.loc[index, "label"] = 1 #좋아요 > 싫어요 -> trust -> label을 1로 설정
    elif(df.loc[index, "sympathy"] < df.loc[index, "unsympathy"]):
        df.loc[index, "label"] = 0 # trust을 제외한 나머지 -> untrust -> label을 0으로 설정
    else:
        df.loc[index, "label"] = -1 # sympathy와 unsympathy가 같은 경우는 뺌. 같은 경우가 많아서 untrust로 넣으면 원하는 결과가 안나올까봐..

t = len(df[df["label"] == 1])
ut = len(df[df["label"] == 0])
       
li = []

n = 0
for i in range(0, t):
    li.append(1)
for i in range(0, ut):
    li.append(0)
df["correlation"] = df["correlation"].abs()
df = df.sort_values(by=['correlation'], axis=0, ascending=False) 
df = df.reset_index()
# correlation score을 기준으로 오름차순, 위에서 t번째까지는 pred가 1, 나머지는 untrust이므로 0개 
df["pred"] = li # column 추가
df_t1 = df[df["pred"] == 1] # trust
df_ut1 = df[df["pred"] == 0] # distrust
df1 = df_t1[['review','rating std','sentiment std','correlation']]
df2 = df_ut1[['review','rating std','sentiment std','correlation']]



df1.to_csv('./definition/definition3_gongjo_t.txt',encoding="utf-8-sig",header=False,index=False)
df2.to_csv('./definition/definition3_gongjo_ut.txt',encoding="utf-8-sig",header=False,index=False) 

```
